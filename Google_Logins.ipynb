```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load dataset from HTML file
df= pd.read_html("C:/Users/jimmy/Documents/Data Projects/Project/jimmyuchuvo.SubscriberInfo.html")
```

```python
#data type from file
type(df)
```

```python
#select the data from list
df1= df[0]
df1
```

```python
#show data info
df1.info()
```

```python
#describe the variables
df1.describe()
```

```python
# set to datetime 
df1.Timestamp= pd.to_datetime(df1.Timestamp)
```

```python
#rename columns
df1= df1.rename(columns={'Timestamp': 'Date', 'Activity Type' :'Login/logout', 'Geo': 'Location'}) 
```

```python
#filter to view Date and Challenges columns when is not null
df1[df1['Date'].notnull() & df1['Challenges'].notnull()].head(5)
```

```python
#rename column
df1= df1.rename(columns= {'Challenges (timestamp, outcome, dusi)': 'Challenges'})
```

```python
df1
```

```python
#split columns into country, province and city 
df1[['Country', 'Province', 'City']]= df1.Location.str.split(',', expand=True)
```

```python
df1
```

```python
#remove the columns that I dont need
df1= df1.drop(columns=['Location','Province'])
```

```python
df1
```

```python
#distribution plot of the login/logout 
print(df1['Login/logout'].value_counts())
print('percent',df1['Login/logout'].value_counts(normalize=True))
sns.displot(df1['Login/logout'])
plt.xticks(rotation=90)
plt.show()
```

```python
#distribution in date
sns.displot(df1['Date'])
plt.xticks(rotation=90)
plt.show()
```

```python
#login/logout distribution by city
sns.countplot(x='City',data= df1, hue='City', palette= 'Set2')
plt.xticks(rotation=90)
plt.show()
```

```python
#distribution by date and city
print(df1.City.value_counts())
print(df1.City.value_counts(normalize=True))
sns.scatterplot(x= df1.Date, y= df1.City)
plt.xticks(rotation=90)
plt.show()
```

```python
# create a column with the name of the day
df1['day_of_week']= df1.Date.dt.day_name()
```

```python
# create a column with the time of the day
bins= [0,12,18,24]
labels= ['Morning', 'Afternoon', 'Evening']
df1['Time_of_the_day']= pd.cut(df1.Date.dt.hour, bins=bins, labels=labels, right= False)
```

```python
#distribution by  day of the week
print(df1.day_of_week.value_counts(normalize= True))
sns.countplot(x= df1.day_of_week, data= df1, color= 'green')
plt.xticks(rotation= 60)
plt.show()
```

```python
#distribution by time of the day
print(df1.Time_of_the_day.value_counts(normalize= True))
sns.displot(df1.Time_of_the_day, color= 'green')
plt.xticks(rotation= 60)
plt.show
```
